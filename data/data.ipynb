```python
import pandas as pd
import yfinance as yf
```

```python
df = pd.read_csv('_BTC-USD_1min.csv', header=0, skiprows=list(range(1, 9)))
```

```python
df
```

```python
df['datetime'] = pd.to_datetime(df.Timestamp, unit='s')
```

```python
df = df.set_index('datetime').asfreq(
    freq='1H').loc['2014-01-01':,][['Open', 'High', 'Low', 'Close', 'Volume_(BTC)']]
```

```python
df
```

```python
df.columns.values[-1] = 'Volume'
```

```python
df
```

```python
df_2 = pd.read_csv('2021.csv')
```

```python
df_2['datetime'] = pd.to_datetime(df_2.unix, unit='s')
```

```python
df_2 = (df_2.set_index('datetime').loc['2021-03-31 01:00:00':'2021-12',]).asfreq(freq='1H')[['open', 'high', 'low', 'close', 'Volume BTC']]
```

```python
df_2.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
```

```python
df
```

```python
df_3 = yf.download('BTC-USD', interval='60m', period='2y').loc['2022':, df.columns.values]
```

```python
df_final = pd.concat([df, df_2, df_3])
```
